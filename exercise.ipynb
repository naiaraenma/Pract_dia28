## Analysis finite size in T

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

**Function with the Gaussian distribution**

It needs the thermostat temperature (Ttermo) and the number of degrees of freedom (g)

```python
def tempinst(T,Tav,sigma):
    f = (1./(np.sqrt(2.*np.pi)*sigma))*np.exp(-(T-Tav)*(T-Tav)/(2.*sigma*sigma))
    return f
    #Crearem una destribució gaussiana, on ens generara temperatures aleatories, Tav i sigma aquests són els tres parametres de la formula
```

**Enter data for the example**

```python
Ttermo=float(input("\n Temperature thermostat (K):\n>")) #temperatura que volem posar
```

```python
g=int(input("\n Number of Degrees of freedom:\n>"))
```

```python
sigma=np.sqrt(2.*Ttermo*Ttermo/g)
print('\nSigma due to finite size of system:',round(sigma,1),' K\n')
```

**Create a list of values of T with its probability and make a plot**

```python
tmin=Ttermo-4.0*sigma
tmax=Ttermo+4.0*sigma
T=np.arange(tmin,tmax,0.005)

plt.xlabel('T (K)')
plt.ylabel('Prob. density')
plt.plot(T,tempinst(T,Ttermo,sigma))
plt.show()
```

**We can generate an expected sequence of measurements**

```python
example_T= np.random.normal(Ttermo,sigma,1000) #generem una seqüencia de numeros o mesures hipotetiques seguint aquesta distribució
```

```python
plt.xlabel('time')
plt.ylabel('Measured T(K)')
plt.plot(example_T)
plt.show()
```
